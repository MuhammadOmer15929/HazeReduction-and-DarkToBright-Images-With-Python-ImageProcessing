# Task#01

```python
# import cv2
import matplotlib.pyplot as plt
import image_dehazer

# Load the input image
haze = cv2.imread("foggyroad.jpg")

# Remove haze from the image using image_dehazer
dehaze, _ = image_dehazer.remove_haze(haze)

# Convert BGR to RGB color space
haze = cv2.cvtColor(haze, cv2.COLOR_BGR2RGB)
dehaze = cv2.cvtColor(dehaze, cv2.COLOR_BGR2RGB)

# Display the input and dehazed images side-by-side using Matplotlib
fig, axs = plt.subplots(1, 2, figsize=(12, 8))
axs[0].imshow(haze)
axs[0].set_title('Input Image')
axs[1].imshow(dehaze)
axs[1].set_title('Dehazed Image')
plt.show()
```

# Task#04

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
dark_image = cv2.imread('boston night.jpg')

# Increase the brightness of the image by adding 50 to each pixel value
bright_image = dark_image + 60

# Display the original and brighter images using matplotlib
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(dark_image, cv2.COLOR_BGR2RGB))
ax[0].set_title('Dark Image')
ax[1].imshow(cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB))
ax[1].set_title('Bright Image')

plt.show()
```
